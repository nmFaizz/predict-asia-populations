```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_rows = 30

df = pd.read_csv('world_population.csv')
df
```

```python
df.info()
```

```python
df.plot()
```

```python
# filtered data frame from continent asia only
filtered_df = df[df['Continent'] == 'Asia']
filtered_df
```

```python
data = ([1970,filtered_df['1970 Population'].sum(axis=0)], 
        [1980,filtered_df['1980 Population'].sum(axis=0)], 
        [1990,filtered_df['1990 Population'].sum(axis=0)],
        [2000,filtered_df['2000 Population'].sum(axis=0)],
        [2010,filtered_df['2010 Population'].sum(axis=0)], 
        [2015,filtered_df['2015 Population'].sum(axis=0)], 
        [2020,filtered_df['2020 Population'].sum(axis=0)], 
        [2022,filtered_df['2022 Population'].sum(axis=0)])

asia_populations = pd.DataFrame(data, columns=['Year', 'Populations'])
asia_populations
```

```python
asia_populations.plot(kind='scatter', x='Year', y='Populations')
plt.title('Year vs Populations')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = asia_populations[['Year']]
y = asia_populations['Populations']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
plt.scatter(x_train, y_train)
plt.title('Year vs Populations')
plt.show()
```

```python
y_predict = model.predict(x_test)

plt.scatter(x_test, y_predict, color='red')
plt.title("Total Populasi Terhadap Tahun")
plt.xlabel("Tahun")
plt.ylabel("Populasi")
plt.show()
```

```python
plt.scatter(x_test, y_test, color='green')
plt.title("Total Populasi Terhadap Tahun")
plt.xlabel("Tahun")
plt.ylabel("Populasi")
plt.show()
```

```python
# comparison data test vs data predict
value = pd.DataFrame(y_test)
value["y_predict"] = y_predict

value
```

```python
# performance of the predict (close to 1 is good performance)
from sklearn.metrics import explained_variance_score
variance = explained_variance_score(y_test,y_predict)
print(variance)
```
